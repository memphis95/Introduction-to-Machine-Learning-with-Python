```python
from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
# transpose the image matrix
img_t = img.T
print(img_t)
```

```python
# reshape
img_reshape = img.reshape(5,20)
print(img_reshape)
```

```python
# sort
img_srt = img.copy()
img_srt.sort(axis = 0)
print(img_srt)
```

```python
# image compression
img_cmp = img.copy()
img_cmp = img_cmp.compress([True, False, True, 0,1,1,1,0,0,1], axis = 0)
print(img_cmp)
```

```python
plt.imshow(img, cmap='Greys')
```

```python
img = imread('astronaut.png')
```

```python
plt.imshow(img)
```

```python
# cut an image slice

img_slice = img.copy()
img_slice = img_slice[0:300, 350:480]
plt.figure()
plt.imshow(img_slice)
```

```python
#  On the Red component of the rocket, 
# all values falling in the range of [100, 150] must be equated to 0
img_slice[np.greater_equal(img_slice[:,:,0],100) & np.less_equal(img_slice[:,:,0],150)] = 0
plt.figure()
plt.imshow(img_slice)
```

```python
img[0:300,350:480,:] = img_slice
```

```python
plt.imshow(img)
```

```python
imf_flip = np.fliplr(img)
```

```python
plt.imshow(imf_flip)
```
